## Import libraries 

```python
from textblob import TextBlob
import matplotlib.pyplot as plt
```

## Run Analysis

```python
with open("20th_century_events.txt", "r", encoding="utf-8") as f:
    text = f.read()

blob = TextBlob(text)
sentiment = blob.sentiment
print(sentiment)  # shows polarity and subjectivity
```

## Create Visualization

```python
labels = ['Polarity', 'Subjectivity']
values = [sentiment.polarity, sentiment.subjectivity]

plt.bar(labels, values, color=['skyblue', 'salmon'])
plt.title("Sentiment Analysis of 20th-Century Events")
plt.ylim(-1, 1)
plt.show() 
```

### The sentiment polarity was close to 0, suggesting neutral overall tone, which was surprising given the number of tragic events listed. Subjectivity was moderately high, possibly due to descriptive phrasing in the events.
